# Maximum Likelihood Estimation 

[![Binder](https://mybinder.org/badge.svg)](https://mybinder.org/v2/gh/danhtaihoang/network-inference/master?filepath=sphinx%2Fcodesource%2Fmle.ipynb)

Maximum likelihood estimation (MLE) is a standard method in the field of network inference. The algorithm of MLE for the kinetic model contains the following steps:

(i) Initialize $W_{ij}$ at random;

(ii) Compute local field $H_i(t) = \sum_j W_{ij} \sigma_j (t)$;

(iii) Compute probability;
$$P[\sigma_i(t+1)|\vec{\sigma}(t)] = \frac{\exp [ \sigma_i(t+1) H_i(t)]}{\sum_{\sigma_i(t+1)} \exp[\sigma_i(t+1) H_i(t)]}$$

(iv) Compute likelihood
$\mathcal{P}_i = \prod_{t=1}^{L-1}P[\sigma_i(t+1)|\vec{\sigma}(t)]$;

(v) Update $W_{ij}$ incrementally through
$W_{ij}^{\textrm{new}} = W_{ij} + \frac{\alpha}{L-1} \frac{\partial \log {\cal{P}}_i }{\partial W_{ij}}$
where $\alpha$ represent learning rate, and
$\frac{\partial \log {\mathcal{P}}_i}{\partial W_{ij}} = \sum_{t=1}^{L-1} \lbrace \sigma_i(t+1) \sigma_j(t) - \tanh [H_i(t)] \sigma_j(t) \rbrace$;

(vi) Repeat (ii)-(v) for some iterations;

(vii) Perform (ii)-(vi) in parallel for every variable index
$i \in \{1, 2, \cdots, N\}$.

In the following, we will test the performance of this method for various cases, small coupling variability and large coupling variability. We will show that the MLE becomes over fitting in the regime of large coupling variability and small sample sizes. Applying our stopping criterion to MLE can make this method works significantly faster and slighlty more accurate.

First of all, we import the packages to the notebook:

```python
import numpy as np
import sys
import timeit
import matplotlib.pyplot as plt
import simulate
import inference
%matplotlib inline

np.random.seed(1)
```

We first consider a system of $N = 100$ variables. Data length $L=2000$ is used for the illustration. Leaning rate $\alpha=1$ is selected.

```python
# parameter setting:
n = 100
l = 2000
rate = 1.
```

## Small coupling variability

For the regime of small coupling variability, as other methods, we select $g = 2.0$ for the demonstration.

```python
g = 2.0
```

The coupling matrix is generated acorrding to the parameters.

```python
w0 = np.random.normal(0.0,g/np.sqrt(n),size=(n,n))
```

From the coupling matrix, we generate time series of variables based on the kinetic ising model:

```python
s = simulate.generate_data(w0,l)
```

Similar to FEM and eMF, MLE is an iterative algorithm. We first run this method as discribed in the literature, without stopping criterion.

```python
start_time = timeit.default_timer()

w = inference.mle(s,rate,stop_criterion='no')

stop_time=timeit.default_timer()
run_time=stop_time-start_time
print('run_time:',run_time)
```

The inferred couplings are plotted to compare with actual couplings:

```python
plt.figure(figsize=(11,3.2))

plt.subplot2grid((1,3),(0,0))
plt.title('actual coupling matrix')
plt.imshow(w0,cmap='rainbow',origin='lower')
plt.xlabel('j')
plt.ylabel('i')
plt.clim(-0.5,0.5)
plt.colorbar(fraction=0.045, pad=0.05,ticks=[-0.5,0,0.5])

plt.subplot2grid((1,3),(0,1))
plt.title('predicted coupling matrix')
plt.imshow(w,cmap='rainbow',origin='lower')
plt.xlabel('j')
plt.ylabel('i')
plt.clim(-0.5,0.5)
plt.colorbar(fraction=0.045, pad=0.05,ticks=[-0.5,0,0.5])

plt.subplot2grid((1,3),(0,2))
plt.title('predicted couplings vs. actual couplings')
plt.plot([-1,1],[-1,1],'r--')
plt.scatter(w0,w)
plt.xlabel('actual couplings')
plt.ylabel('inferred couplings')

plt.tight_layout(h_pad=1, w_pad=1.5)
plt.show()
```

Mean square error is calculated as the following:

```python
MSE = np.mean((w-w0)**2)
print('MSE:',MSE)
```

Now, we apply our stopping criterion for this method. According to this criterion, we stop the iteration when the discrepancy between observed variables and its model expectations starts to increase.

```python
start_time = timeit.default_timer()

w = inference.mle(s,rate,stop_criterion='yes')

stop_time=timeit.default_timer()
run_time=stop_time-start_time
print('run_time:',run_time)
```

The inferred couplings are plotted:

```python
plt.figure(figsize=(11,3.2))

plt.subplot2grid((1,3),(0,0))
plt.title('actual coupling matrix')
plt.imshow(w0,cmap='rainbow',origin='lower')
plt.xlabel('j')
plt.ylabel('i')
plt.clim(-0.5,0.5)
plt.colorbar(fraction=0.045, pad=0.05,ticks=[-0.5,0,0.5])

plt.subplot2grid((1,3),(0,1))
plt.title('predicted coupling matrix')
plt.imshow(w,cmap='rainbow',origin='lower')
plt.xlabel('j')
plt.ylabel('i')
plt.clim(-0.5,0.5)
plt.colorbar(fraction=0.045, pad=0.05,ticks=[-0.5,0,0.5])

plt.subplot2grid((1,3),(0,2))
plt.title('predicted couplings vs. actual couplings')
plt.plot([-1,1],[-1,1],'r--')
plt.scatter(w0,w)
plt.xlabel('actual couplings')
plt.ylabel('inferred couplings')

plt.tight_layout(h_pad=1, w_pad=1.5)
plt.show()
```

```python
MSE = np.mean((w-w0)**2)
print('MSE:',MSE)
```

So, our stopping criterion makes MLE works slightly better (in accuracy) and significantly faster.

## Large coupling variability

```python
g = 4.0
w0 = np.random.normal(0.0,g/np.sqrt(n),size=(n,n))

s = simulate.generate_data(w0,l)
```
